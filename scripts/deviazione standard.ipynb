```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers= ['PG', 'BEI.DE']
sec_data = pd.DataFrame()

for t in tickers:
    sec_data[t]= wb.DataReader(t, data_source='yahoo', start='2007-1-1')['Adj Close']
```

```python
sec_data.tail()
```

```python
sec_returns=np.log(sec_data/sec_data.shift(1))
sec_returns
```

```python
#calcolo il ritorno medio delle security
sec_returns['PG'].mean()
```

```python
sec_returns['PG'].mean()*250
```

```python
sec_returns['PG'].std()
#deviazione standard
```

```python
#dobbiamo ora calcolare la radice di 250 e moltiplicarlo per la std
sec_returns['PG'].std()*250**0.5
```

```python
#posso velocizzare il procedimento con più titoli, occhio alle parentesi
sec_returns[['PG','BEI.DE']].mean()*250
```

```python
sec_returns[['PG','BEI.DE']].std() *25 **0.5
```
