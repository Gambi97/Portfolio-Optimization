```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers= ['PG', 'BEI.DE','AAPL']
sec_data = pd.DataFrame()

for t in tickers:
    sec_data[t]= wb.DataReader(t, data_source='yahoo', start='2007-1-1')['Adj Close']
    
```

```python
sec_returns=np.log(sec_data/sec_data.shift(1))
sec_returns
```

```python
weights= np.array([0.333, 0.333 , 0.333])
```

```python
#calcolo della varianza del portafoglio
pfoglio_var= np.dot(weights.T, np.dot(sec_returns.cov()*250, weights))
pfoglio_var
```

```python
pfoglio_vol= np.dot(weights.T, np.dot(sec_returns.cov()*250, weights))**0.5
pfoglio_vol
```

```python
print (str(round(pfoglio_vol,5)*100)+ ' %')
```

```python
#sistematico e non sistematico gestione del portafoglio
PG_var_a = sec_returns['PG'].var()*250
BEI_var_a = sec_returns['BEI.DE'].var()*250
AAPL_var_a = sec_returns['AAPL'].var()*250

#diversifiable
dr= pfoglio_var - (weights[0] **2 *PG_var_a) - (weights[1] **2 *BEI_var_a)  - (weights[2] **2 *AAPL_var_a)
dr
```

```python
print (str(round(dr*100, 3))+ ' %')
```

```python
n_dr= pfoglio_var - dr
n_dr
```

```python
print (str(round(n_dr*100, 3))+ ' %')
```
