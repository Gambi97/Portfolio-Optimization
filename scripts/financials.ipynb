funzione che estrare e ordina le info più interessanti del profile delle aziende 

```python
#Profile con dati che interessano
def profile_company(assets):
    #La funzione ha bisogno in ingresso di una lista tipo questa:
    #assets = ['INTC','IBM', 'MET','SBUX','WMT','AAPL']
    index = ['price', 'beta', 'volAvg', 'mktCap', 'lastDiv', 'range', 'changes',
           'changesPercentage', 'companyName', 'exchange', 'industry', 'website',
           'description', 'ceo', 'sector', 'image']
    profile = pd.DataFrame(index = index)

    for ticker in assets:
        profile[ticker] = fa.profile(ticker)

    profile = profile.loc[['companyName','sector','industry','mktCap','beta'],:]
    return profile
```
