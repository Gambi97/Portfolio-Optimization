```python
#log return è meglio quando dobbiamo calcolare il ritorno di un sola zion
import numpy as np
import pandas as pd
```

```python
from pandas_datareader import data as wb
```

```python
PG = wb.DataReader('PG', data_source='yahoo', start='2009-1-1')
```

```python
PG
```

```python
#ritorno logaritmico 
PG['log_return']= np.log(PG['Adj Close']/ PG['Adj Close'].shift(1))
print(PG['log_return'])
```

```python
#per stampare in un piano il rendimento logaritmico delle azioni
PG['log_return'].plot(figsize=(9,6))
```

```python
#calcolo il tirorno medio percentuale giornaliero
daily_return= PG['log_return'].mean()
#stampo a video
daily_return
```

```python
#calcolo il ritorno annuo 
annual_return=PG['log_return'].mean()*250
#250 vsono i giorni in cui è aperto il mercato
annual_return
```

```python
#lo stampo a percentuale
str(round(annual_return,3)*100) +'%'
#str per convertitlo a stringa e aggiungere quindi poter aggiungere %
```
