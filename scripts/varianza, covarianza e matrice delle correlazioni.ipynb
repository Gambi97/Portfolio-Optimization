```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers= ['PG', 'BEI.DE','AAPL']
sec_data = pd.DataFrame()

for t in tickers:
    sec_data[t]= wb.DataReader(t, data_source='yahoo', start='2015-1-1')['Adj Close']
```

```python
sec_returns=np.log(sec_data/sec_data.shift(1))
sec_returns
```

```python
#calcolo della varianza di una singola azione
PG_var= sec_returns['PG'].var()
PG_var
```

```python
PG_var_a= sec_returns['PG'].var()*250
PG_var_a
```

```python
BEI_var_a= sec_returns['BEI.DE'].var()*250
BEI_var_a
```

```python
#calcolo della covarianza delle due azioni, sulla diagonale le singole azioni, sulla parte 
#triangolare la loro covarianza
#con questa matrice posso bypassare i calcoli della varianza iniziale
cov_matrix_a= sec_returns.cov()*250
cov_matrix_a
```

```python
#ora calcoliamo la loro correlazione
corr_matrix= sec_returns.corr()
corr_matrix
```
