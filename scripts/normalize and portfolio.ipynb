```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
azioni = ['PG', 'MSFT', 'F', 'GE']
dati=pd.DataFrame()
for i in azioni:
    dati[i]=wb.DataReader(i, data_source='yahoo', start='2005-1-1')['Adj Close']
```
